```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('/Users/kanshruti/Downloads/CA Fires.csv')

shape = df.shape

describe = df.describe()

data_one = df.query('YEAR > 1900 and GIS_ACRES > 100000')

data_two = df.loc[:, "CAUSE"]

print(describe)

```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,4))

ax1 = fig.add_subplot(121)

ax1.set_xlabel("YEAR")

ax1.set_ylabel("CAUSE")

ax1.set_title("Cause by Year")

data_one.plot(kind='bar')
```

```python
fig, ax = plt.subplots() 

data = df['YEAR'].value_counts() 

year = data.index 
frequency = data.values 

ax.set_xlim(1950,2020)

ax.bar(year, frequency) 

ax.set_title('Rate of Fires') 
ax.set_xlabel('Year') 
ax.set_ylabel('Frequency')
```
